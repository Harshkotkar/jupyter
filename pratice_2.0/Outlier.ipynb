# outlier
An outlier is a data point that significantly differs from the other values in a dataset.These extreme values can distort analyses, models, and predictions.

Types of Outliers:
Univariate Outliers: These are outliers in a single variable (e.g., a single column in a dataset).
Multivariate Outliers: These occur when an observation differs from the general distribution in multiple variables at once.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
house=pd.DataFrame({"price":[10,13,16,19,23,26,20,27,98]})
house.head(3)
```

```python
house.describe()
```

```python
sns.boxplot(x="price",data=house)
plt.show()
```

```python
sns.distplot(house["price"])
plt.show()
```

```python
q1=house['price'].quantile(0.25)
q3=house['price'].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
min_range=q1-(1.5*IQR)
max_range=q3+(1.5*IQR)
min_range,max_range
```

```python
df=house[house["price"]<=max_range]
```

```python
sns.boxplot(x="price",data=df)
plt.show()
```
