```python

name=data['Short_name'].head(20)
fare=data['Fare'].head(20)
fig,ax=plt.subplots(figsize=(20,10))
ax.barh(name,fare,color='skyblue')

for s in ['top','bottom','left','right']:
    ax.spines[s].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

ax.xaxis.set_tick_params(pad=5)
ax.yaxis.set_tick_params(pad=10)

ax.grid(visible=True,color='grey',
       linestyle='-.',linewidth=0.5,alpha=0.2)
for i in ax.patches:
    plt.text(i.get_width()+0.2,i.get_y()+0.5, 
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')

ax.set_title('Titanic name and fare',
             loc ='left', )

fig.text(0.9, 0.15," ", fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)
plt.show()
```

```python
#using matplotlib

year=[2014,2015,2016,2017,2018,2019,2020,2021,2022,2023,2024]#bunny
incomes=[38,46,55,56,46,61,73,49,40,34,84]
income=[33,36,45,56,76,67,87,69,76,72,74]#cassendra

plt.plot(year,incomes,linewidth=0.6)
plt.plot(year,income,linewidth=0.6)

income_ticks = list(range(30, 90, 5))
plt.yticks(income_ticks, [f"{x}k USD" for x in income_ticks])

plt.xlabel("Years")
plt.ylabel(" Income in USD")
plt.legend(['bunny','cassendra'])
plt.xticks(year,rotation=45)
plt.grid(False)
plt.show()
```

```python

#using seabon
import seaborn as sns


year = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
b_incomes = [38, 46, 55, 56, 46, 61, 73, 49, 40, 34, 84]
c_incomes = [33, 36, 45, 56, 76, 67, 87, 69, 76, 72, 74]

df = pd.DataFrame({"year": year, "b_income": b_incomes, "c_income": c_incomes})


plt.figure(figsize=(20,10))
sns.lineplot(x='year', y='b_income', data=df, marker='o', label='Bunny Income')
sns.lineplot(x='year', y='c_income', data=df, marker='o', label='Cassendra Income')


plt.title('Income Comparison of Bunny and Cassendra ')
plt.xlabel('Year')
plt.ylabel('Income')
plt.grid(False)
plt.legend(loc="upper left")
plt.xticks(year, rotation=45)
plt.tight_layout()
plt.show()
```

```python
#subplots

x=np.arange(50)

fig,axs=plt.subplots(2,2)

axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title("Sine wave")

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title('Cosine wave')

axs[1,0].plot(x,np.random.random(50))
axs[1,0].set_title("Random values")

axs[1,1].plot(x,np.log(x))
axs[1,1].set_title("Log Function")

plt.show()
```
