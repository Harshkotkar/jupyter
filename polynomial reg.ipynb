```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1.5, 3.6, 7.9, 15.1, 25.3])

# Transform features to polynomial
degree = 2  # Change degree for experimentation
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)

# Fit the model
model = LinearRegression()
model.fit(X_poly, y)

# Make predictions
y_pred = model.predict(X_poly)

# Plot
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', label=f'Polynomial Degree {degree}')
plt.legend()
plt.title('Polynomial Regression')
plt.show()

# Model performance
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5,8,10,9,5,7,7.8,9.2,14,12.3,9.8,12.4]).reshape(-1, 1)
y = np.array([1.5, 3.7, 8.4, 14.9, 25.1,23.4,30,28,29.6,9,21,6,10,17.8,32,15.4])

# Polynomial transformation (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the Linear Regression model
model = LinearRegression()
model.fit(X_poly, y)

# Predictions
y_pred = model.predict(X_poly)

# Visualize
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', label='Polynomial Fit')
plt.legend()
plt.title("Polynomial Regression")
plt.show()

# Print model coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
r2_score = model.score(X_poly, y)
print(r2_score)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Generate 200 data points
np.random.seed(0)  # For reproducibility
X = np.linspace(0, 10, 200).reshape(-1, 1)
# Create a non-linear relationship (example: y = 2x^2 + 3x + noise)
y = 2 * X**2 + 3 * X + np.random.normal(0, 5, X.shape)

# Polynomial transformation (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the Linear Regression model
model = LinearRegression()
model.fit(X_poly, y)

# Predict for a custom value (e.g., X = 9.5)
custom_X = 10
predicted_y = model.predict(poly.transform([[custom_X]]))

# Print the predicted result for X = 9.5
print(f"Predicted y value for X = {custom_X}: {predicted_y[0]}")

# Visualize
y_pred = model.predict(X_poly)
plt.scatter(X, y, color='blue', s=10, label='Original Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Polynomial Fit')
plt.legend()
plt.title("Polynomial Regression (200 Points)")
plt.xlabel("X")
plt.ylabel("y")
plt.show()

# Print model coefficients and R² value
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
r2_score = model.score(X_poly, y)
print(f"R² value: {r2_score}")
```
