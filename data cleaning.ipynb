```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()/df.shape[0]*100 #percentage null value present in every column
```

```python
df.isnull().sum().sum() #total null value present in the dataset
```

```python
(df.isnull().sum().sum()/(df.shape[0]*df.shape[1]))*100 #total percentage null value present in the dataset
```

```python
sns.heatmap(df.isnull()) #
```

```python
sns.displot(
    data=df.isnull().melt(value_name='missing'),
    y='variable',
    hue='missing',
    multiple='fill',
    height=8,
    aspect=1.1
)

# specifying a threshold value
plt.axvline(0.4, color='r')
```
